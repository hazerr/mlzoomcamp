## Enunciado del problema
Cómo puede un banco predecir la probabilidad de que un cliente abandone la empresa, y desarrollar programas de fidelización y campañas de retención para reducir la tasa de abandono?

Descripción del problema:

Los bancos tienen un interés en retener a sus clientes existentes, ya que es mucho más caro adquirir nuevos clientes que mantener los existentes. Para ello, es importante que los bancos puedan predecir la probabilidad de que un cliente abandone la empresa.

La predicción de la probabilidad de abandono puede hacerse mediante el análisis de datos de clientes, como datos demográficos, financieros y de comportamiento. Una vez que los bancos tienen una comprensión de los factores que llevan a los clientes a abandonar la empresa, pueden desarrollar programas de fidelización y campañas de retención para reducir la tasa de abandono.

Objetivos del problema:

El objetivo de este problema es desarrollar un modelo que pueda predecir la probabilidad de que un cliente abandone un banco. Este modelo se utilizará para desarrollar programas de fidelización y campañas de retención para reducir la tasa de abandono.

Enfoques para resolver el problema:

Existen varios enfoques para resolver este problema, como:

Modelos de aprendizaje automático: Los modelos de aprendizaje automático pueden utilizarse para aprender de los datos de clientes y predecir la probabilidad de abandono.
Análisis de datos: El análisis de datos puede utilizarse para identificar los factores que llevan a los clientes a abandonar la empresa.
Beneficios de resolver el problema:

La resolución de este problema puede generar los siguientes beneficios para los bancos:

Reducción de la tasa de abandono: Al reducir la tasa de abandono, los bancos pueden conservar a sus clientes existentes y ahorrar dinero en gastos de adquisición de nuevos clientes.
Aumento de la satisfacción del cliente: Los programas de fidelización y las campañas de retención pueden ayudar a mejorar la satisfacción del cliente y aumentar la lealtad a la marca.
Mejora de la rentabilidad: Al retener a los clientes existentes, los bancos pueden generar más ingresos y mejorar su rentabilidad.

# 3. Entrenar modelo

```python
#df = pd.get_dummies(df)
df = pd.get_dummies(df)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = df.drop('Exited', axis=1)
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, df['Exited'], test_size=0.25, random_state=42)
```

boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(learning_rate=0.01, n_estimators=100, max_depth=5)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
score = clf.score(X_test, y_test)

print('Accuracy:', score)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, max_depth=5)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
score = clf.score(X_test, y_test)

print('Accuracy:', score)
```

arboles

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
score = clf.score(X_test, y_test)

print('Accuracy:', score)
```

random forest 1

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
score = clf.score(X_test, y_test)

print('Accuracy:', score)
```

```python
from xgboost import XGBClassifier

clf = XGBClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
score = clf.score(X_test, y_test)

print('Accuracy:', score)
```

```python
#mejora de hiperparametros
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5]
}

grid_search = GridSearchCV(clf, param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_clf = grid_search.best_estimator_

print('Best parameters:', best_params)
print('Accuracy:', best_clf.score(X_test, y_test))
```

```python
from xgboost import XGBClassifier

clf = XGBClassifier(max_depth=15, min_samples_leaf=1, min_samples_split=10, n_estimators=10)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
score = clf.score(X_test, y_test)

print('Accuracy:', score)
```
